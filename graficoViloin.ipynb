```python
# Importando as Bibliotecas usadas no projeto
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.datasets import load_iris
```

```python
# Carregando o dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
```

```python
# Visualizando as primeiras linhas do DataFrame
df.head()
```

```python
# Explorando as Estatísticas Descritivas
df.describe()
```

```python
# Visualizando a Distribuição das Variáveis
for coluna in df.columns[:-1]:
  plt.figure(figsize=(10, 6))
  plt.hist(df[coluna], bins=10, color='skyblue', edgecolor='black')
  plt.grid(True)
  plt.xlabel(coluna)
  plt.ylabel('Frequência')
  plt.title(f"Histograma de {coluna}")
  plt.show()
  print("\n")
```

```python
# Gerando os gráficos de Violino

# Usando Matplotlib
plt.figure(figsize=(10, 6))
plt.violinplot([df[df['species'] == especie]['sepal length (cm)'] for especie in df['species'].unique()],
               showmeans=True, showmedians=True, showextrema=True)
plt.xticks([1, 2, 3], df['species'].unique())
plt.xlabel('Espécie')
plt.ylabel('Comprimento da Sépala (cm)')
plt.title('Gráfico de Violino - Matplotlib')
plt.show()
```

```python
# Gerando os gráficos de Violino

# Usando Seaborn
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal length (cm)', data=df)
plt.xlabel('Espécie')
plt.ylabel('Comprimento da Sépala (cm)')
plt.title('Gráfico de Violino - Seaborn')
plt.show()
```

```python
# Gerando os gráficos de Violino

# Usando Plotly
fig = px.violin(df, x='species', y='sepal length (cm)', box=True, points='all',
                color='species', color_discrete_map={
                    'setosa': '#17BECF',
                    'versicolor': '#FFA500',
                    'virginica': '#800080'
                })
fig.update_layout(
    xaxis_title='Espécie',
    yaxis_title='Comprimento da Sépala (cm)',
    title='Gráfico de Violino - Plotly'
)
fig.show()
```
